*Array Reshaping , Ravel , Flatten*

```python
arr1 = np.array([1,2,3,4,5,6])
arr2 = arr1.reshape(2,3)
```

```python
arr1
```

```python
arr2
```

```python
arr3 = arr2.reshape(3,2)
arr3
```

```python
ravel = arr3.ravel()
ravel
```

```python
ravel[0] = 100
ravel[1] = 200
ravel
```

```python
arr3
```

```python
arr4 = np.array([3,4,5,6,7,8])
arr5 = arr4.reshape(2,3)
```

```python
arr5
```

```python
flat = arr5.flatten()
flat
```

```python
flat[0] = 300
flat
```

```python
arr5
```

```python
arr = np.array([10,20,30,40,50])
ind = [0,2]
print(np.take(arr,ind))
```

```python
for x in np.nditer(arr):
  print(x , end= " ")
```

```python
array = np.array([[1,2,3] , [3,4,6]])
```

```python
for ind , x in np.ndenumerate(array):
  print(ind, x)
```
